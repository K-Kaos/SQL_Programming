```python
import pandas as pd
column_name=['year', 'genre', 'publisher', 'total_sales']

game_trend_year=pd.read_csv('./Dataset/game_trend_year.csv', names = column_name)
game_trend_year.head(20)
```

```python
import matplotlib.pyplot as plt

#   
yearly_sales = game_trend_year.groupby('year')['total_sales'].sum()

#    
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Yearly Game Sales Trend')
plt.grid(True)
plt.show()
#2005 2010    
```

```python
import matplotlib.pyplot as plt

# 2005~2010 
filtered_df = game_trend_year[(game_trend_year['year'] >= 2005) & (game_trend_year['year'] <= 2010)]

# 2005~2010 trend game genre 
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

years = filtered_df['year'].unique()

for i, year in enumerate(years):
    ax = axs[i//3][i%3] 
    
    year_data = filtered_df[filtered_df['year'] == year]
    
    ax.bar(year_data['genre'], year_data['total_sales'])
    ax.set_title(f'Year {year} Game Sales')
    ax.set_xlabel('Genre')
    ax.set_ylabel('Total Sales')
    ax.tick_params(axis='x', rotation=90)  

plt.tight_layout() 
plt.show()
```

```python
column_name=['platform','year', 'genre', 'total_sales']
game_trend_platform=pd.read_csv('./Dataset/game_trend_platform.csv', names = column_name)
game_trend_platform.head(20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


#   
platforms = game_trend_platform['platform'].unique()

#       
for platform in platforms:
    platform_data = game_trend_platform[game_trend_platform['platform'] == platform]
    years = platform_data['year']
    total_sales = platform_data['total_sales']
    plt.plot(years, total_sales, label=platform)

#     
plt.title('Game Sales Trend by Platform')
plt.xlabel('Year')
plt.ylabel('Total Sales')

#  
plt.legend()

#  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


#  
platform_groups = game_trend_platform.groupby('platform')

#  
fig, axs = plt.subplots(len(platform_groups), figsize=(10, 6 * len(platform_groups)))

#     
for i, (platform, data) in enumerate(platform_groups):
    ax = axs[i] if len(platform_groups) > 1 else axs
    data.plot(x='year', y='total_sales', kind='line', ax=ax)
    ax.set_title(f'Platform: {platform}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Total Sales')

#  
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
column_names = ['platform', 'year', 'genre', 'total_sales']
game_trend_platform = pd.read_csv('./Dataset/game_trend_platform.csv', names=column_names)

#    
platform_generations = {
    '3DS': '7th',
    'DC': '6th',
    'GB': '4th',
    'GBA': '6th',
    'GC': '6th',
    'N64': '5th',
    'NES': '3rd',
    'PC': 'Varies',
    'PS': '5th',
    'PS2': '6th',
    'PS3': '7th',
    'PS4': '8th',
    'PSP': '7th',
    'PSV': '8th',
    'SNES': '4th',
    'Wii': '7th',
    'WiiU': '8th',
    'X360': '7th',
    'XB': '6th',
    'XOne': '8th'
}

#   
generation_groups = game_trend_platform.groupby(game_trend_platform['platform'].map(platform_generations))

#  
fig, axs = plt.subplots(len(generation_groups), figsize=(10, 6 * len(generation_groups)))

#  
colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown']

#     
for i, (generation, data) in enumerate(generation_groups):
    ax = axs[i] if len(generation_groups) > 1 else axs
    for platform, color in zip(data['platform'].unique(), colors):
        platform_data = data[data['platform'] == platform]
        ax.plot(platform_data['year'], platform_data['total_sales'], label=platform, color=color)
    ax.set_title(f'Generation: {generation}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Total Sales')
    ax.legend()

#  
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
column_names = ['platform', 'year', 'genre', 'total_sales']
game_trend_platform = pd.read_csv('./Dataset/game_trend_platform.csv', names=column_names)

#     
platform_genre_counts = game_trend_platform.groupby(['platform', 'genre']).size().unstack(fill_value=0)

#  
fig, ax = plt.subplots(figsize=(12, 8))

# Heatmap 
im = ax.imshow(platform_genre_counts, cmap='YlGnBu')

#  
ax.set_xticks(range(len(platform_genre_counts.columns)))
ax.set_yticks(range(len(platform_genre_counts.index)))
ax.set_xticklabels(platform_genre_counts.columns, rotation=90)
ax.set_yticklabels(platform_genre_counts.index)

#   
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Count', rotation=-90, va="bottom")

#     
ax.set_title('Platform Preference by Genre')
ax.set_xlabel('Genre')
ax.set_ylabel('Platform')

#  
plt.tight_layout()
plt.show()
```
